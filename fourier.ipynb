```python
import matplotlib.pyplot as plt
import numpy as np
```

## Função

$$
x(t) = e^{at}u(t)
$$

```python
def func(x, a):
    return np.e**(-a*x)
```

```python
a = 5
x_values = np.linspace(0, 50, 10)
```

```python
y_values = func(x_values, a)
plt.plot(x_values, y_values)
plt.xlabel(r'$\omega$')
plt.ylabel(r'$x(t)$')
plt.grid()
```

## Amplitude do sinal de fourier
$$
 |X(\omega) | =  \frac{1}{\sqrt{a^2 + \omega^2}}
$$

com $\omega$ variando $-\infty$ a $\infty$

```python
def amplitude(x:list, a:int):
    return 1 / np.sqrt(a**2 + x**2)
```

```python
y_values = amplitude(x_values, a)
plt.plot(x_values, y_values)
plt.xlabel(r'$\omega$')
plt.ylabel(r'$|X(\omega)|$')
plt.grid()
```

## Fase da Série e transformação de Fourier

$$
\angle{H(\omega) = \tan^{-1}(\frac{\omega}{a}})
$$

```python
def fase(x:list, a:int):
    return -np.arctan(x/a)
```

```python
a = 5
x_values = np.linspace(0, 10, 1000)
y_values = fase(x_values, a)
plt.plot(x_values, y_values)
plt.xlabel(r'$\omega$')
plt.ylabel(r'$\angle{H(\omega)}$')
plt.grid()
```
